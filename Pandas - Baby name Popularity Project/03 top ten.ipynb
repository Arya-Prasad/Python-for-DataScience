## To extract the most popular names in any given year

1. Sorting a dataframe  by a column
2. Dropping columns from a Dataframe
3. Joining frames based on index
4. Counting values in a Series


```python
#Select all data for a year using a loc selection object

allyears_indexed.loc['M',:,2008].head()
```

```python
#Next sort this selection with larger numbers on top

allyears_indexed.loc['M',:,2008].sort_values('number',ascending = False).head()
```

The most popular boy name in 2008 was Jacob

```python
#Assign this to a table by feeding this code to pop2008
 
pop2008 = allyears_indexed.loc['M',:,2008].sort_values('number',ascending = False).head()

```

```python
#Reset the index and drop columns that are not required

pop2008.reset_index().drop(['sex','year','number'],axis=1).head()
```

```python
#Encapsulate these manipulations into a function so that we can copy
#the top ten for any year.

def topten(sex,year):
    simple = allyears_indexed.loc[sex,:,year].sort_values('number',ascending=False).reset_index()
    simple = simple.drop(['sex','year','number'],axis=1).head(10)
    
    simple.columns = [year]  #change name of one column to year
    simple.index = simple.index+1
    
    return simple
```

```python
topten('M',2009)
```

```python
#To compare several years this way compute another function that takes
#a range of years and compute top tens for each year and then will use
#pandas function join to put them in a single dataframe.

def toptens(sex,year0,year1):
    years = [topten(sex,year) for year in range(year0,year1+1)]
    
    return years[0].join(years[1:])

```

```python
toptens('M',2000,2010)
```

```python
toptens('F',1985,1995)
```

```python
#Take the six most popular names in this decade and plot 
#a popularity across all years.

#To see the rise and fall and eventual defeat we turn this dataframe 
#into a series by stacking it

toptens('F',1985,1995).stack().head()
```

```python
#to compute counts of all the values 

toptens('F',1985,1995).stack().value_counts()
```

```python
#Plot the first six, need to grab the values from the index of this series

popular = toptens('F',1985,1995).stack().value_counts().index[:6]
```

```python
pp.figure(figsize=(12,2.5))

for name in popular:
    plotname('F',name)
    
pp.legend(popular)
```
