```python
def parsefile(filename):
    return np.genfromtxt(filename,
                         delimiter = dly_delimiter,
                         usecols = dly_usecols,
                         dtype = dly_dtype,
                         names = dly_names)
```

```python
dly_delimiter = [11,4,2,4] + [5,1,1,1] * 31
dly_usecols = [1,2,3] + [4*i for i in range(1,32)]
dly_dtype = [np.int32,np.int32,(np.str_,4)] + [np.int32] * 31
dly_names = ['year','month','obs'] + [str(day) for day in range(1,31+1)]
```

```python
def unroll(record):
    startdate = np.datetime64('{}-{:02}'.format(record['year'],record['month']))
    dates = np.arange(startdate,startdate + np.timedelta64(1,'M'),np.timedelta64(1,'D'))
    
    rows = [(date,record[str(i+1)]/10) for i,date in enumerate(dates)]
    
    return np.array(rows,dtype=[('date','M8[D]'),('value','d')])
```

```python
def getobs(filename,obs):
    return np.concatenate([unroll(row) for row in parsefile(filename) if row[2] == obs])
```

```python
def getobs(filename,obs):
    data = np.concatenate([unroll(row) for row in parsefile(filename) if row[2] == obs])
    
    data['value'][data['value'] == -999.9] = np.nan
    
    return data
```

## Smoothing data

 Although we have now filled in the missing data, our temperature series are still very noisy. To see trends on longer time scales we need to average out short-term fluctuations. This is known technically, as smoothing.
 
1. Smoothing data with a running mean in numpy
2. Placing subplots together with matplotlib

```python
#Looking at one of the plots

pp.plot(lihue_tmin['date'],lihue_tmin['value'])
```

Indeed, it's very noisy

```python
#To get a better result, we will smooth out short-term oscillations.
#by taking a running mean, that is a mean over a limited window centered at the datapoint.

def plot_smoothed(t,win=10):
    smoothed = np.correlate(t['value'],np.ones(win)/win,'same')
    pp.plot(t['date'],smoothed)
    
#win --- size of the window
#np.correlate --- multiplies the sliding section of a 1D array with another shorter array, 
#                 computes the sum of the multiplied values and stores that in a new array.
#This is just what we need for the running mean if we set the shorter array to a 1/size.
#The 'same' option make sure  that the size of the resulting array is same as the original one.
```

```python
plot_smoothed(lihue_tmin)
```

```python
#looking at a smaller slice of this array and different smoothing parameters.
pp.plot(lihue_tmin[10000:12000]['date'],lihue_tmin[10000:12000]['value'])  #to compare with original plot

plot_smoothed(lihue_tmin[10000:12000])
plot_smoothed(lihue_tmin[10000:12000],30)
```

Here we can see, indeed, smoothing in action. At the edges we see artifacts due to the fact that the smoothing window runs out of values to average on the left or the right.

```python
#Plotting all stations together
#we kept the codes for the stations in the list datastations.
#We'll call plot_smoothed on the arrays that we get for each of them. 
#We give the right file name to get obs, call it for TMIN and for TMAX, 
#add a very large smoothing parameter(365)

for i,code in enumerate(datastations):
    
    pp.subplot(2,2,i+1)  #To place all plots in different subplots within a larger plot
    
    plot_smoothed(getobs('{}.dly'.format(code),'TMIN'),365)
    plot_smoothed(getobs('{}.dly'.format(code),'TMAX'),365)
```

```python
#Improving the plots

pp.figure(figsize=(10,6))   #to make the plot bigger

for i,code in enumerate(datastations):
    
    pp.subplot(2,2,i+1)  #To place all plots in different subplots within a larger plot
    
    plot_smoothed(getobs('{}.dly'.format(code),'TMIN'),365)
    plot_smoothed(getobs('{}.dly'.format(code),'TMAX'),365)

    pp.title(stations[code])
    pp.axis(xmin=np.datetime64('1952'),xmax=np.datetime64('2012'),ymin=-10,ymax=30)
    
pp.tight_layout()  #To apply a tight layout
```
