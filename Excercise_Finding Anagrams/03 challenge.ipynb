```python
import collections
```

```python
wordclean = sorted(list(set([word.strip().lower() for word in open('words','r')])))
```

```python
def signature(word):
    return ''.join(sorted(word))
```

```python
words_bysig = collections.defaultdict(list)

for word in wordclean:
    words_bysig[signature(word)].append(word)
```

```python
def anagram_fast(myword):
    return words_bysig[signature(myword)]
```

```python
anagrams_all = {word: anagram_fast(word) for word in wordclean if len(anagram_fast(word)) > 1}
```

```python
words_bylen = collections.defaultdict(list)

for word in wordclean:
    words_bylen[len(word)].append(word)
```

```python
anagram_bylen = {}

for length,words in words_bylen.items():
    anagram_bylen[length] = {word: anagram_fast(word) for word in words if len(anagram_fast(word)) > 1}
```

```python
# To get the number of anagrams of each word

anagram_bylen = {}

for length,words in words_bylen.items():
    anagram_bylen[length] = {word: len(anagram_fast(word)) for word in words if len(anagram_fast(word)) > 1}
```

```python
anagram_bylen
```

```python
#To not include the word itself in the count

anagram_bylen = {}

for length,words in words_bylen.items():
    anagram_bylen[length] = {word: len(anagram_fast(word))-1 for word in words if len(anagram_fast(word)) > 1}
```

```python
anagram_bylen
```

```python
# To count number of anagrams in each class

anagram_bylen = {}

for length,words in words_bylen.items():
    anagram_bylen[length] = sum(len(anagram_fast(word))-1 for word in words if len(anagram_fast(word)) > 1)/2
```

We want to sum all of these lengths. We realize also that we should divide the total number of anagrams by two. That is because each anagram is being counted twice. For instance, the anagram "Elvis lives" will be counted once when we meet the word "Elvis" and once when we meet the word "lives".
