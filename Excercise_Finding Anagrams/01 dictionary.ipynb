_Finding anagrams in the dictionary of English words._

_Also includes intro to a few useful Python idioms. An example may be loading words from a text file into a Python list. So we are first going to see how to load an English dictionary into Python, then we are going to create a Python dict of anagrams. _



```python
word = open('words','r')
```

```python
word  #word is a file object
```

```python
#readlines method on the file object returns a list of all the lines in the file

wordlist = word.readlines()
```

```python
wordlist[:10]
```

```python
len(wordlist)
```

```python
# To get rid of newline characters at the end of each string and convert all the words to lowercase

'Aaron\n'.strip()
```

```python
'Aaron'.lower()
```

```python
wordclean = [word.strip().lower() for word in wordlist]
#uses list comprehension and chaining the methods strip and lower
```

```python
wordclean[:10]
```

```python
#To remove duplicate words create a python 'set' structure that holds only unique elements and then convert back to a list

wordunique = list(set(wordclean))
```

```python
wordunique
```

```python
# To sort it in alphabetical order again

wordunique.sort()
wordunique[:10]
```

```python
# It is possible to do all these reading and parsing steps in asingle step with comprehension

wordunique = sorted(list(set([word.strip().lower() for word in open('words','r')])))
```

```python
wordunique[:10]
```

_Using a list comprehension to go through the lines of a file, as we just did, is one example of a Python idiom. These are language constructs that encapsulate a specially convenient or expressive solutions to common problems. _
